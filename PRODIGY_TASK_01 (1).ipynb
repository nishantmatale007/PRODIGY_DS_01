```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
import os
```

```python
# To load data
data=pd.read_csv("IMDB-Movie-Data.csv.xls")
```

```python
data
```

```python
df = pd.DataFrame(data)
```

```python
Title = df['Title'].head(12)
Rating = df ['Rating'].head(12)
```

```python
# Figure size
fig = plt.figure(figsize=(10,7))
```

```python
# Horizontal bar plot
plt.bar(Title [0:10], Rating[0:10])
plt.xticks(rotation=90)
```

```python
# dataframe for ratings of movies
ratings = df ['Rating']
```

```python
# Plotting histogram
plt.hist(ratings, bins=10, color='skyblue', edgecolor='black')
# Adding Labels and title
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('IMDb Movie Ratings Distribution')
```

```python
# Display the plot
plt.show()
```

```python
#Horizontal Bar Plot:
import matplotlib.pyplot as plt
df= pd.DataFrame(data)
Title = df['Title'].head(12)
Rating = df ['Rating'].head(12)
# Create horizontal bar plot
plt.figure(figsize=(10, 6)) # Adjust figure size if needed
plt.barh(Title [0:12], Rating [0:12], color='pink')
plt.xlabel('IMDb Rating')
plt.title('Movies IMDb Rating')
plt.gca().invert_yaxis() # Invert y-axis to display the highest-rated movie on top
plt.show()
```

```python
# dataframe for movies released in year 2007
movies_2007 = data[data['Year'] == 2007]
movies_2007
```

```python
#Vertical Bar Plot:
Title = movies_2007['Title'].head(15)
Revenue = movies_2007['Revenue (Millions)'].head(15)
plt.bar(Title, Revenue)
plt.title('Revenue of few movies released in 2016')
plt.xlabel('Movie Name')
plt.ylabel('Revenue (Millions)')
plt.xticks (rotation=90)
plt.show()
```

```python
#dataframe for highest rated movies by IMDB
movies_ratings = data[data['Rating'] >= 8.5]
movies_ratings
```

```python
#dataframe for ratings of movies 
highest_rated_movies = movies_ratings['Rating']
```

```python
# Plotting histogram
plt.hist(highest_rated_movies, bins=6, color='red', edgecolor="black")
# Adding Labels and title
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Highest movies rated by IMdB')
```
